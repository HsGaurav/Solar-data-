```python
import numpy as np
import matplotlib.pyplot as plt

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

# plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
```

```python
#CURVE OF BEST FIT FOR POWER OF FIXED MOUNT IN SINGLE AXIS
import numpy as np
import matplotlib.pyplot as plt

w0 = np.arange(8.0, 19.0, 1.0)
w1 = np.array([20.664, 39.780, 44.176, 70.616, 88.110, 104.960, 125.334, 105.342, 86.172, 70.620, 46.494])

curve_fit = np.polyfit(w0, w1, 3)
curve = np.poly1d(curve_fit)
plt.plot(w0, w1, 'o', list(range(7, 20)), curve(list(range(7, 20))))
grid = (True)
plt.xlabel('Time')
plt.ylabel('Output Power (mW)')
```
